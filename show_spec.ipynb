```python
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt 

```

```python
k = 3

ok_dataset = os.listdir("/home/jongwook95.lee/study/cnn_based_audio_classification/dataset/audio_files/test/ok/")
ok_file = os.path.join("/home/jongwook95.lee/study/cnn_based_audio_classification/dataset/audio_files/test/ok/", ok_dataset[k])
```

```python
x = librosa.load(ok_file,16000)[0]
y = librosa.stft(x, n_fft=1024, hop_length=256, win_length=128)

magnitude = np.abs(y)
log_spectrogram = librosa.amplitude_to_db(magnitude)

plt.figure(figsize=(10,4))
librosa.display.specshow(log_spectrogram, sr=16000, hop_length=256)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar(format='%+2.0f dB')
plt.title("Spectrogram (dB)")
```
